```python
# Cell: Instance‑Based Training & Evaluation with Logging & Saving

import os, time, random, numpy as np, pandas as pd
import torch, torch.nn as nn, torch.optim as optim
from collections import deque

# ─── 0. Force CPU & Setup ─────────────────────────────────────────────────
os.environ["CUDA_VISIBLE_DEVICES"] = ""
os.environ["TORCH_COMPILE_DISABLE"] = "1"
device = torch.device("cpu")

seed = 0
random.seed(seed); np.random.seed(seed)
torch.manual_seed(seed)

# ─── Hyperparameters ───────────────────────────────────────────────────────
hidden_layers = [512, 256, 128]
lr            = 1e-4
gamma         = 0.99
batch_size    = 64
eps_start     = 1.0
eps_end       = 0.05
eps_decay     = 0.995
sync_steps    = 500
episodes      = 5000
log_int       = 1000  # print every log_int episodes

# ─── 1. Data load ─────────────────────────────────────────────────────────
def load_tsp20(path):
    coords_list, tours = [], []
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            idx = parts.index("output")
            vals = list(map(float, parts[:idx]))
            tour = list(map(int, parts[idx+1:]))
            coords_list.append(np.array(vals, dtype=np.float32).reshape(-1,2))
            tours.append(tour)
    return coords_list, tours

coords_list, tours = load_tsp20("data/homework/hw_2_tsp_20.txt")
N = coords_list[0].shape[0]

# ─── 2. Env, Net, Buffer ───────────────────────────────────────────────────
class TSPEnv:
    def __init__(self, coords):
        self.coords_base = coords; self.N = coords.shape[0]
    def reset(self):
        self.coords = self.coords_base.copy()
        self.visited = np.zeros(self.N, bool)
        self.current = 0; self.visited[0] = True
        return self._state()
    def _state(self):
        return np.concatenate([self.coords.flatten(),
                               self.visited.astype(np.float32),
                               np.eye(self.N, dtype=np.float32)[self.current]])
    def step(self, a):
        prev, self.current = self.current, a
        self.visited[a] = True
        d = np.linalg.norm(self.coords[prev] - self.coords[a])
        r = -d
        done = self.visited.all()
        if done:
            # return-to-start
            r -= np.linalg.norm(self.coords[self.current] - self.coords[0])
        return self._state(), r, done, {}

class DQNNet(nn.Module):
    def __init__(self, s_dim, a_dim, hidden):
        super().__init__()
        layers, in_dim = [], s_dim
        for h in hidden:
            layers += [nn.Linear(in_dim, h), nn.ReLU()]; in_dim = h
        layers.append(nn.Linear(in_dim, a_dim))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, cap=20000): self.buf = deque(maxlen=cap)
    def push(self, *exp): self.buf.append(exp)
    def sample(self, B):
        batch = random.sample(self.buf, B)
        s,a,r,s2,d = zip(*batch)
        return (
            torch.from_numpy(np.stack(s)).float().to(device),
            torch.tensor(a,      dtype=torch.long,   device=device),
            torch.tensor(r,      dtype=torch.float32,device=device),
            torch.from_numpy(np.stack(s2)).float().to(device),
            torch.tensor(d,      dtype=torch.float32,device=device)
        )
    def __len__(self): return len(self.buf)

class DQNAgent:
    def __init__(self, s_dim, a_dim):
        self.gamma = gamma; self.N = a_dim
        self.policy = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target.load_state_dict(self.policy.state_dict())
        self.opt = optim.Adam(self.policy.parameters(), lr=lr)
        self.buf = ReplayBuffer()
        self.loss_fn = nn.SmoothL1Loss()

    def select(self, state, eps):
        visited = state[2*self.N : 3*self.N]
        unv = np.where(visited < 0.5)[0]
        if random.random() < eps:
            return int(np.random.choice(unv))
        st = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            qvals = self.policy(st)[0].cpu().numpy()
        qvals[visited > 0.5] = -np.inf
        return int(np.argmax(qvals))

    def update(self, B):
        if len(self.buf) < B: return
        s,a,r,s2,d = self.buf.sample(B)
        q_s = self.policy(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            a_max = self.policy(s2).argmax(1, keepdim=True)
            q2    = self.target(s2).gather(1, a_max).squeeze(1)
            y     = r + self.gamma * q2 * (1 - d)
        loss = self.loss_fn(q_s, y)
        self.opt.zero_grad(); loss.backward(); self.opt.step()

    def sync(self): self.target.load_state_dict(self.policy.state_dict())

# ─── 3. Training & Evaluation with Logging ─────────────────────────────────
all_R, all_T = [], []
for i in range(5):
    coords, tour = coords_list[i], tours[i]
    env   = TSPEnv(coords)
    agent = DQNAgent(4*N, N)
    eps, rewards, step = eps_start, [], 0
    t0 = time.time()

    for ep in range(1, episodes+1):
        s, done, tr = env.reset(), False, 0.0
        while not done:
            a = agent.select(s, eps)
            s2, r, done, _ = env.step(a)
            agent.buf.push(s, a, r, s2, float(done))
            s, tr = s2, tr + r
            agent.update(batch_size); step += 1
            if step % sync_steps == 0:
                agent.sync()

        rewards.append(tr)
        eps = max(eps_end, eps * eps_decay)

        # --- Logging at intervals ---
        if ep == 1 or ep % log_int == 0:
            elapsed = time.time() - t0
            avg100 = np.mean(rewards[-100:]) if len(rewards) >= 100 else rewards[-1]
            print(f"[Inst {i+1}] Ep {ep}/{episodes}  ε={eps:.3f}  avg100={avg100:.2f}  time={elapsed:.1f}s")

    total_t = time.time() - t0
    all_R.append(rewards); all_T.append(total_t)

    # --- Test evaluation and print as before ---
    opt_nodes = [n-1 for n in tour]
    opt_cost = sum(np.linalg.norm(coords[u]-coords[v]) for u,v in zip(opt_nodes[:-1],opt_nodes[1:]))
    s, done, dqn = env.reset(), False, [0]
    while not done:
        a = agent.select(s, eps=0.0)
        s, _, done, _ = env.step(a)
        dqn.append(a)
    cycle = dqn + [dqn[0]]
    dqn_cost = sum(np.linalg.norm(coords[u]-coords[v]) for u,v in zip(cycle[:-1],cycle[1:]))
    print(f"\n== Test {i+1} ==")
    print(f" Optimal route = {tour}")
    print(f" DQN route     = {[n+1 for n in cycle]}")
    print(f" Optimal cost  = {opt_cost:.4f}")
    print(f" DQN cost      = {dqn_cost:.4f}")
    print(f" Gap (%)       = {(dqn_cost - opt_cost)/opt_cost*100:.2f}%")

# ─── 4. Save Experiment Results ───────────────────────────────────────────
os.makedirs("results", exist_ok=True)
for idx, R in enumerate(all_R, start=1):
    np.save(f"results/rewards_instance_{idx}.npy", np.array(R, dtype=np.float32))
summary = [{
    "instance": idx,
    "training_time_s": T,
    "final_reward": float(R[-1]),
    "mean_reward_last100": float(np.mean(R[-100:])) if len(R)>=100 else float(np.mean(R))
} for idx, (R, T) in enumerate(zip(all_R, all_T), start=1)]
pd.DataFrame(summary).to_csv("results/summary.csv", index=False)
print("✔ All instance results saved to 'results/'")
```

### Distribution-based

```python
# ─── Cell: Distribution‑Based DQN Training & Evaluation with Test Printing ───
import os, time, random, glob
import numpy as np
import pandas as pd
import torch, torch.nn as nn, torch.optim as optim
from collections import deque

# ─── 0. Setup ─────────────────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
seed = 0
random.seed(seed); np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

# ─── Hyperparameters ───────────────────────────────────────────────────────
hidden_layers = [512, 512]   # MLP: 4N → 512 → 512 → N
lr            = 1e-4         # learning rate
gamma         = 0.99         # discount factor
batch_size    = 64           # replay batch size
eps_start     = 1.0          # initial ε
eps_end       = 0.05         # minimum ε
eps_decay     = 0.995        # ε decay per episode
sync_steps    = 500          # target network sync frequency
episodes      = 5000         # total distribution episodes
log_int       = 1000         # print every 1000 eps

# ─── 1. Data load ─────────────────────────────────────────────────────────
# 1.1 500 training instances
train_files  = sorted(glob.glob("data/TSP20/instances/instance_*.npz"))
train_coords = [np.load(f)["coords"].astype(np.float32) for f in train_files]

# 1.2 5 test instances from txt
def load_test(path):
    coords_list, tours = [], []
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            i = parts.index("output")
            vals = list(map(float, parts[:i]))
            tour = list(map(int,   parts[i+1:]))
            coords_list.append(np.array(vals, np.float32).reshape(-1,2))
            tours.append(tour)
    return coords_list, tours

test_coords, test_tours = load_test("data/homework/hw_2_tsp_20.txt")
N = train_coords[0].shape[0]

# ─── 2. Environment ────────────────────────────────────────────────────────
class TSPEnv:
    def __init__(self, coords):
        self.coords_base = coords; self.N = coords.shape[0]
    def reset(self):
        self.coords = self.coords_base.copy()
        self.visited = np.zeros(self.N, bool)
        self.current = 0; self.visited[0] = True
        return self._state()
    def _state(self):
        return np.concatenate([
            self.coords.flatten(),                      # 2N
            self.visited.astype(np.float32),            # N
            np.eye(self.N, dtype=np.float32)[self.current]  # N
        ])
    def step(self, a):
        prev = self.current; self.current = a; self.visited[a] = True
        d = np.linalg.norm(self.coords[prev] - self.coords[a])
        r = -d
        done = self.visited.all()
        if done:
            # 돌아가기 보상
            r -= np.linalg.norm(self.coords[self.current] - self.coords[0])
        return self._state(), r, done, {}

# ─── 3. Network & ReplayBuffer ────────────────────────────────────────────
class DQNNet(nn.Module):
    def __init__(self, s_dim, a_dim, hidden):
        super().__init__()
        layers, in_dim = [], s_dim
        for h in hidden:
            layers += [nn.Linear(in_dim, h), nn.ReLU()]; in_dim = h
        layers.append(nn.Linear(in_dim, a_dim))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, cap=20000): self.buf = deque(maxlen=cap)
    def push(self, *exp):       self.buf.append(exp)
    def sample(self, B):
        batch = random.sample(self.buf, B)
        s,a,r,s2,d = zip(*batch)
        return (
            torch.from_numpy(np.stack(s)).float().to(device),
            torch.tensor(a, dtype=torch.long,   device=device),
            torch.tensor(r, dtype=torch.float32,device=device),
            torch.from_numpy(np.stack(s2)).float().to(device),
            torch.tensor(d, dtype=torch.float32,device=device),
        )
    def __len__(self): return len(self.buf)

# ─── 4. Agent ─────────────────────────────────────────────────────────────
class DQNAgent:
    def __init__(self, s_dim, a_dim):
        self.gamma = gamma; self.N = a_dim
        self.policy = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target.load_state_dict(self.policy.state_dict())
        self.opt = optim.Adam(self.policy.parameters(), lr=lr)
        self.buf = ReplayBuffer()
        self.loss_fn = nn.SmoothL1Loss()

    def select(self, state, eps):
        visited = state[2*self.N:3*self.N]
        unv = np.where(visited < 0.5)[0]
        if random.random() < eps:
            return int(np.random.choice(unv))
        st = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            qvs = self.policy(st)[0].cpu().numpy()
        qvs[visited > 0.5] = -np.inf
        return int(np.argmax(qvs))

    def update(self, B):
        if len(self.buf) < B: return
        s,a,r,s2,d = self.buf.sample(B)
        q_s = self.policy(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            a_max = self.policy(s2).argmax(1, keepdim=True)
            q2    = self.target(s2).gather(1, a_max).squeeze(1)
            y     = r + self.gamma * q2 * (1 - d)
        loss = self.loss_fn(q_s, y)
        self.opt.zero_grad(); loss.backward(); self.opt.step()

    def sync(self):
        self.target.load_state_dict(self.policy.state_dict())

# ─── 5. Distribution‑Based Training ────────────────────────────────────────
agent_dist = DQNAgent(4*N, N)
eps = eps_start
dist_rewards = []
step = 0
t0 = time.time()

for ep in range(1, episodes+1):
    idx = random.randrange(len(train_coords))
    env = TSPEnv(train_coords[idx])
    s, done, tr = env.reset(), False, 0.0

    while not done:
        a = agent_dist.select(s, eps)
        s2, r, done, _ = env.step(a)
        agent_dist.buf.push(s, a, r, s2, float(done))
        s, tr = s2, tr + r
        agent_dist.update(batch_size); step += 1
        if step % sync_steps == 0:
            agent_dist.sync()

    dist_rewards.append(tr)
    eps = max(eps_end, eps * eps_decay)

    if ep == 1 or ep % log_int == 0:
        elapsed = time.time() - t0
        avg100 = np.mean(dist_rewards[-100:]) if len(dist_rewards)>=100 else tr
        print(f"[Dist] Ep {ep}/{episodes}  ε={eps:.3f}  avg100={avg100:.2f}  time={elapsed:.1f}s")

total_t_dist = time.time() - t0
print(f"Distribution training completed in {total_t_dist:.1f}s")

# ─── 6. Evaluation on 5 Test Instances ───────────────────────────────────
for i, coords in enumerate(test_coords, start=1):
    tour = test_tours[i-1]
    env = TSPEnv(coords)

    # optimal cost
    opt_nodes = [n-1 for n in tour]
    opt_cost  = sum(np.linalg.norm(coords[u]-coords[v])
                    for u,v in zip(opt_nodes[:-1], opt_nodes[1:]))

    # DQN route
    s, done, dqn = env.reset(), False, [0]
    while not done:
        a = agent_dist.select(s, eps=0.0)
        s, _, done, _ = env.step(a)
        dqn.append(a)

    cycle = dqn + [dqn[0]]
    dqn_cost = sum(np.linalg.norm(coords[u]-coords[v])
                   for u,v in zip(cycle[:-1], cycle[1:]))

    print(f"\n== Test {i} ==")
    print(f" Optimal route = {tour}")
    print(f" DQN route     = {[n+1 for n in cycle]}")
    print(f" Optimal cost  = {opt_cost:.4f}")
    print(f" DQN cost      = {dqn_cost:.4f}")
    print(f" Gap (%)       = {(dqn_cost-opt_cost)/opt_cost*100:.2f}%")

# ─── 7. Save Distribution Results ─────────────────────────────────────────
os.makedirs("results", exist_ok=True)
np.save("results/rewards_distribution.npy",
        np.array(dist_rewards, dtype=np.float32))

summary_dist = {
    "mode": "distribution",
    "training_time_s": total_t_dist,
    "final_reward": float(dist_rewards[-1]),
    "mean_reward_last100": float(np.mean(dist_rewards[-100:]))
}
pd.DataFrame([summary_dist]).to_csv(
    "results/summary_distribution.csv", index=False
)
print("✔ Distribution results saved to 'results/'")
```

### hybrid

```python
# Cell: Hybrid Instance & Distribution-Based DQN Training & Evaluation

import os, time, random, glob
import numpy as np
import pandas as pd
import torch, torch.nn as nn, torch.optim as optim
from collections import deque

# ─── 0. Setup ─────────────────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
seed = 0
random.seed(seed); np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

# ─── Hyperparameters ───────────────────────────────────────────────────────
hidden_layers = [512, 512]
lr            = 1e-4
gamma         = 0.99
batch_size    = 64
eps_start     = 1.0
eps_end       = 0.05
eps_decay     = 0.995
sync_steps    = 500
episodes      = 5000
log_int       = 1000
p_instance    = 0.5   # 혼합 확률: instance-based 수행 확률

# ─── 1. Data load ─────────────────────────────────────────────────────────
# 1.1 500 training instances (Distribution pool)
train_files  = sorted(glob.glob("data/TSP20/instances/instance_*.npz"))
train_coords = [np.load(f)["coords"].astype(np.float32) for f in train_files]

# 1.2 5 test instances (and for instance-based sampling)
def load_test(path):
    coords_list, tours = [], []
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            i = parts.index("output")
            vals = list(map(float, parts[:i]))
            tour = list(map(int,   parts[i+1:]))
            coords_list.append(np.array(vals, np.float32).reshape(-1,2))
            tours.append(tour)
    return coords_list, tours

test_coords, test_tours = load_test("data/homework/hw_2_tsp_20.txt")
N = train_coords[0].shape[0]

# ─── 2. Environment ────────────────────────────────────────────────────────
class TSPEnv:
    def __init__(self, coords):
        self.coords_base = coords; self.N = coords.shape[0]
    def reset(self):
        self.coords = self.coords_base.copy()
        self.visited = np.zeros(self.N, bool)
        self.current = 0; self.visited[0] = True
        return self._state()
    def _state(self):
        return np.concatenate([
            self.coords.flatten(),
            self.visited.astype(np.float32),
            np.eye(self.N, dtype=np.float32)[self.current]
        ])
    def step(self, a):
        prev = self.current; self.current = a; self.visited[a] = True
        d = np.linalg.norm(self.coords[prev] - self.coords[a])
        r = -d
        done = self.visited.all()
        if done:
            r -= np.linalg.norm(self.coords[self.current] - self.coords[0])
        return self._state(), r, done, {}

# ─── 3. Network & Buffer ─────────────────────────────────────────────────
class DQNNet(nn.Module):
    def __init__(self, s_dim, a_dim, hidden):
        super().__init__()
        layers, in_dim = [], s_dim
        for h in hidden:
            layers += [nn.Linear(in_dim, h), nn.ReLU()]; in_dim = h
        layers.append(nn.Linear(in_dim, a_dim))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, cap=20000): self.buf = deque(maxlen=cap)
    def push(self, *exp): self.buf.append(exp)
    def sample(self, B):
        batch = random.sample(self.buf, B)
        s,a,r,s2,d = zip(*batch)
        return (
            torch.from_numpy(np.stack(s)).float().to(device),
            torch.tensor(a, dtype=torch.long,   device=device),
            torch.tensor(r, dtype=torch.float32,device=device),
            torch.from_numpy(np.stack(s2)).float().to(device),
            torch.tensor(d, dtype=torch.float32,device=device)
        )
    def __len__(self): return len(self.buf)

# ─── 4. Agent ─────────────────────────────────────────────────────────────
class DQNAgent:
    def __init__(self, s_dim, a_dim):
        self.gamma = gamma; self.N = a_dim
        self.policy = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target = DQNNet(s_dim, a_dim, hidden_layers).to(device)
        self.target.load_state_dict(self.policy.state_dict())
        self.opt = optim.Adam(self.policy.parameters(), lr=lr)
        self.buf = ReplayBuffer()
        self.loss_fn = nn.SmoothL1Loss()
    def select(self, state, eps):
        visited = state[2*self.N:3*self.N]
        unv = np.where(visited < 0.5)[0]
        if random.random() < eps:
            return int(np.random.choice(unv))
        st = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            qvs = self.policy(st)[0].cpu().numpy()
        qvs[visited > 0.5] = -np.inf
        return int(np.argmax(qvs))
    def update(self, B):
        if len(self.buf) < B: return
        s,a,r,s2,d = self.buf.sample(B)
        q_s = self.policy(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            a_max = self.policy(s2).argmax(1, keepdim=True)
            q2    = self.target(s2).gather(1, a_max).squeeze(1)
            y     = r + self.gamma * q2 * (1 - d)
        loss = self.loss_fn(q_s, y)
        self.opt.zero_grad(); loss.backward(); self.opt.step()
    def sync(self): self.target.load_state_dict(self.policy.state_dict())

# ─── 5. Hybrid Training ───────────────────────────────────────────────────
agent = DQNAgent(4*N, N)
eps = eps_start
hybrid_rewards = []
step = 0
t0 = time.time()

for ep in range(1, episodes+1):
    # 믹스: instance‑based vs distribution‑based
    if random.random() < p_instance:
        # instance-based: 무작위로 평가 5개 중 하나 선택
        idx = random.randrange(len(test_coords))
        env = TSPEnv(test_coords[idx])
    else:
        # distribution-based: 500개 훈련 인스턴스 중 하나 선택
        idx = random.randrange(len(train_coords))
        env = TSPEnv(train_coords[idx])
    s, done, tr = env.reset(), False, 0.0

    while not done:
        a = agent.select(s, eps)
        s2, r, done, _ = env.step(a)
        agent.buf.push(s, a, r, s2, float(done))
        s, tr = s2, tr + r
        agent.update(batch_size)
        step += 1
        if step % sync_steps == 0:
            agent.sync()

    hybrid_rewards.append(tr)
    eps = max(eps_end, eps * eps_decay)

    if ep == 1 or ep % log_int == 0:
        elapsed = time.time() - t0
        avg100 = np.mean(hybrid_rewards[-100:]) if len(hybrid_rewards)>=100 else tr
        mode = "Inst" if random.random()<p_instance else "Dist"
        print(f"[Hybrid] Ep {ep}/{episodes} ε={eps:.3f} avg100={avg100:.2f} time={elapsed:.1f}s")

total_t = time.time() - t0
print(f"Hybrid training completed in {total_t:.1f}s")

# ─── 6. Evaluation on 5 Test Instances ───────────────────────────────────
for i, coords in enumerate(test_coords, start=1):
    tour = test_tours[i-1]
    env = TSPEnv(coords)
    # optimal cost
    opt_nodes = [n-1 for n in tour]
    opt_cost  = sum(np.linalg.norm(coords[u]-coords[v])
                    for u,v in zip(opt_nodes[:-1], opt_nodes[1:]))
    # DQN route
    s, done, dqn = env.reset(), False, [0]
    while not done:
        a = agent.select(s, eps=0.0)
        s, _, done, _ = env.step(a)
        dqn.append(a)
    cycle = dqn + [dqn[0]]
    dqn_cost = sum(np.linalg.norm(coords[u]-coords[v])
                   for u,v in zip(cycle[:-1], cycle[1:]))
    print(f"\n== Test {i} ==")
    print(f"Optimal route = {tour}")
    print(f"DQN route     = {[n+1 for n in cycle]}")
    print(f"Optimal cost  = {opt_cost:.4f}")
    print(f"DQN cost      = {dqn_cost:.4f}")
    print(f"Gap (%)       = {(dqn_cost-opt_cost)/opt_cost*100:.2f}%")

# ─── 7. Save Hybrid Results ───────────────────────────────────────────────
os.makedirs("results", exist_ok=True)
np.save("results/rewards_hybrid.npy", np.array(hybrid_rewards, dtype=np.float32))
summary = {
    "mode": "hybrid",
    "training_time_s": total_t,
    "final_reward": float(hybrid_rewards[-1]),
    "mean_reward_last100": float(np.mean(hybrid_rewards[-100:]))
}
pd.DataFrame([summary]).to_csv("results/summary_hybrid.csv", index=False)
print("✔ Hybrid results saved to 'results/'")
```

### 1. Data Generation

```python
import os
import time
import numpy as np
from python_tsp.exact import solve_tsp_dynamic_programming

# 설정
node_counts = [10, 15, 20]
num_samples = 500
seed = 1234

np.random.seed(seed)

for n in node_counts:
    # 폴더 준비
    base_dir = f"data/TSP{n}"
    inst_dir = os.path.join(base_dir, "instances")
    sol_dir  = os.path.join(base_dir, "solutions")
    os.makedirs(inst_dir, exist_ok=True)
    os.makedirs(sol_dir,  exist_ok=True)

    start_all = time.time()
    valid = 0

    while valid < num_samples:
        # 1) 무작위 좌표 생성
        coords = np.random.random((n, 2))  # [0,1]^2

        # 2) 거리 행렬 계산
        dist_matrix = np.linalg.norm(
            coords[:, None, :] - coords[None, :, :],
            axis=2
        )

        # 3) DP 기반 exact TSP 해 계산
        t0 = time.time()
        try:
            tour, cost = solve_tsp_dynamic_programming(dist_matrix)
        except MemoryError:
            # n이 너무 크면 DP가 메모리 부족할 수 있음
            continue
        elapsed = time.time() - t0

        # 4) 정상 투어인 경우 저장
        if tour is not None and len(tour) == n:
            fname = f"{valid:03d}"

            # 좌표 저장
            np.savez_compressed(
                os.path.join(inst_dir, f"instance_{fname}.npz"),
                coords=coords
            )

            # 솔루션(투어, 코스트, 소요 시간) 저장
            np.savez_compressed(
                os.path.join(sol_dir, f"solution_{fname}.npz"),
                tour=tour,
                cost=cost,
                solve_time_s=elapsed
            )

            valid += 1
            print(f"[TSP{n} #{fname}] cost={cost:.2f}, time={elapsed:.2f}s")

#    print(f"TSP{n}: Generated {valid} instances in {time.time() - start_all:.1f}s")
```

### DQN

```python
# Cell 1: Imports & Device 설정
import os, glob
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# Cell 2: 데이터 로드 및 파라미터 정의
instance_files = sorted(glob.glob("data/TSP20/instances/*.npz"))
solution_files = sorted(glob.glob("data/TSP20/solutions/*.npz"))
coords_list = [np.load(f)["coords"].astype(np.float32) for f in instance_files]
opt_costs   = [float(np.load(f)["cost"]) for f in solution_files]
N = coords_list[0].shape[0]

# 학습 설정
# Mode①: 각 인스턴스별 에피소드 수
episodes_per_instance = 1000
# Mode②: 전체 분포 학습 에피소드 수
episodes_distribution = 2000
batch_size = 32
eps_start, eps_end, eps_decay = 1.0, 0.01, 0.995
target_update_freq = 100

print(f"Loaded {len(coords_list)} instances, N={N}")
print(f"Instance mode: {episodes_per_instance} eps per inst  |  Distribution mode: {episodes_distribution} total eps")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TSPEnv:
    def __init__(self, coords, mode="instance"):
        self.base = coords
        self.N = coords.shape[0]
        self.mode = mode
    def reset(self):
        self.coords = self.base.copy() if self.mode=="instance" else np.random.rand(self.N,2).astype(np.float32)
        self.visited = np.zeros(self.N, bool)
        self.current = 0
        self.visited[0] = True
        return self._state()
    def _state(self):
        return np.concatenate([
            self.coords.flatten(),
            self.visited.astype(np.float32),
            np.eye(self.N, dtype=np.float32)[self.current]
        ])
    def step(self, a):
        prev = self.current
        self.current = a
        self.visited[a] = True
        d = np.linalg.norm(self.coords[prev] - self.coords[a])
        r = -d
        done = self.visited.all()
        return self._state(), r, done, {}

class DQNNet(nn.Module):
    def __init__(self, s_dim, a_dim, h=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, h), nn.ReLU(),
            nn.Linear(h, h),     nn.ReLU(),
            nn.Linear(h, a_dim)
        )
    def forward(self, x):
        return self.net(x)

class ReplayBuffer:
    def __init__(self, cap=10000):
        self.cap = cap
        self.buf = []
        self.pos = 0
    def push(self, s, a, r, s2, d):
        if len(self.buf) < self.cap:
            self.buf.append(None)
        self.buf[self.pos] = (s, a, r, s2, d)
        self.pos = (self.pos + 1) % self.cap
    def sample(self, n):
        idx = np.random.choice(len(self.buf), n, replace=False)
        batch = [self.buf[i] for i in idx]
        s, a, r, s2, d = zip(*batch)
        return (
            torch.tensor(s, dtype=torch.float32, device=device),
            torch.tensor(a, dtype=torch.long,    device=device),
            torch.tensor(r, dtype=torch.float32, device=device),
            torch.tensor(s2, dtype=torch.float32, device=device),
            torch.tensor(d, dtype=torch.float32, device=device)
        )
    def __len__(self):
        return len(self.buf)

class DQNAgent:
    def __init__(self, s_dim, a_dim, lr=1e-3, gamma=0.99):
        self.gamma = gamma
        self.policy = DQNNet(s_dim, a_dim).to(device)
        self.target = DQNNet(s_dim, a_dim).to(device)
        self.target.load_state_dict(self.policy.state_dict())
        self.opt = optim.Adam(self.policy.parameters(), lr=lr)
        self.buf = ReplayBuffer()
    def select(self, s, eps):
        if np.random.rand() < eps:
            return np.random.randint(0, self.policy.net[-1].out_features)
        state_t = torch.tensor(s, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            q = self.policy(state_t)
        return int(q.argmax(dim=1))
    def update(self, batch_size):
        if len(self.buf) < batch_size:
            return
        s, a, r, s2, d = self.buf.sample(batch_size)
        q_s = self.policy(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            q2 = self.target(s2).max(1)[0]
            y  = r + self.gamma * q2 * (1 - d)
        loss = nn.functional.mse_loss(q_s, y)
        self.opt.zero_grad()
        loss.backward()
        self.opt.step()
    def sync(self):
        self.target.load_state_dict(self.policy.state_dict())
```

```python
# Cell 4: 학습 함수 (Instance & Distribution) – 시간 측정 & 러닝 커브 기록
def train_instance(agent, coords, episodes, label):
    env=TSPEnv(coords,"instance")
    eps=eps_start
    rewards=[]
    start=time.time()
    for ep in range(1,episodes+1):
        s=env.reset(); done=False; tot=0
        while not done:
            a=agent.select(s,eps); s2,r,done,_=env.step(a)
            agent.buf.push(s,a,r,s2,float(done)); s=s2; agent.update(batch_size); tot+=r
        rewards.append(tot); eps=max(eps_end,eps*eps_decay)
        if ep%target_update_freq==0: agent.sync()
    elapsed=time.time()-start
    print(f"{label} trained in {elapsed:.1f}s")
    return rewards, elapsed

def train_distribution(agent, coords_list, episodes, label):
    eps=eps_start; rewards=[]
    start=time.time()
    for ep in range(1,episodes+1):
        coords=coords_list[np.random.randint(len(coords_list))]
        env=TSPEnv(coords,"instance"); s=env.reset(); done=False; tot=0
        while not done:
            a=agent.select(s,eps); s2,r,done,_=env.step(a)
            agent.buf.push(s,a,r,s2,float(done)); s=s2; agent.update(batch_size); tot+=r
        rewards.append(tot); eps=max(eps_end,eps*eps_decay)
        if ep%target_update_freq==0: agent.sync()
    elapsed=time.time()-start
    print(f"{label} trained in {elapsed:.1f}s")
    return rewards, elapsed
```

```python
import os
import time
import numpy as np

# Cell 5: 학습 실행 & 시간 측정
state_dim = 4*N; action_dim=N
agent1 = DQNAgent(state_dim,action_dim)
rew1, t1 = train_instance(agent1, coords_list[0], episodes_per_instance, "Instance Mode")
agent2 = DQNAgent(state_dim,action_dim)
rew2, t2 = train_distribution(agent2, coords_list, episodes_distribution, "Distribution Mode")
```

```python
# 솔루션과 DQN 결과 간 갭(%) 계산 코드

import numpy as np

def compute_gap(agent, coords_list, opt_costs):
    gaps = []
    for coords, opt in zip(coords_list, opt_costs):
        env = TSPEnv(coords, mode="instance")
        state = env.reset()
        done = False
        tour = [env.current]
        while not done:
            action = agent.select(state, eps=0.0)
            state, _, done, _ = env.step(action)
            tour.append(action)
        # 총 경로 비용 계산
        cost = sum(np.linalg.norm(coords[tour[i]] - coords[tour[i+1]]) 
                   for i in range(len(tour)-1))
        # 상대 갭(%)
        gap = (cost - opt) / opt * 100
        gaps.append(gap)
    gaps = np.array(gaps)
    print(f"Mean gap: {gaps.mean():.2f}%")
    print(f"Std  gap: {gaps.std():.2f}%")
    print(f"Min  gap: {gaps.min():.2f}%")
    print(f"Max  gap: {gaps.max():.2f}%")
    return gaps

# Instance 모드 에이전트 평가
print("=== Instance Mode Gap ===")
gaps_inst = compute_gap(agent1, coords_list, opt_costs)

# Distribution 모드 에이전트 평가
print("\n=== Distribution Mode Gap ===")
gaps_dist = compute_gap(agent2, coords_list, opt_costs)
```
